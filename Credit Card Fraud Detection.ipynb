## Model Building 

```python
from sklearn.pipeline import make_pipeline,Pipeline
from sklearn.model_selection import KFold,StratifiedKFold
from sklearn.model_selection import cross_val_score,cross_val_predict
```

```python
skf=StratifiedKFold(n_splits=4,random_state=1)
```

```python
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.metrics import classification_report,confusion_matrix,roc_auc_score,f1_score
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('Class',axis=1)
y=df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=101)
```

```python
xgb=XGBClassifier(max_depth=1000,n_estimators=500,learning_rate=0.2)
rf=RandomForestClassifier(class_weight='balanced',n_jobs=-1,bootstrap=True,max_depth=1000)
svc=SVC(class_weight='balanced')
extra=ExtraTreesClassifier(class_weight='balanced',n_jobs=-1)
```

## Cross Validation

```python
xgb_score=cross_val_score(xgb,X_train,y_train,scoring='precision',cv=skf,n_jobs=-1,verbose=1,)
```

```python
xgb_score.mean()
```

```python
rf_param_grid= [{'max_depth': [80,500,1000],
                 'max_features':[2,3,4],
                 'n_estimators':[200,300,1000],
                 'class_weight':['balanced']}]
```

```python
grid_search= GridSearchCV(rf,param_grid=rf_param_grid,cv=2,n_jobs=-1,verbose=2)
```

```python
rf_score=cross_val_score(rf,X_train,y_train,scoring='precision',cv=skf,n_jobs=-1,verbose=1)
```

```python
print(rf_score.min())
print(rf_score.mean())
print(rf_score.max())
```

```python
rf_pred=cross_val_predict(rf,X_train,y_train,cv=skf,n_jobs=-1,verbose=1)
```

```python
temp=pd.DataFrame(rf_pred)
```

## Voting Classifier Model 

```python
from sklearn.ensemble import VotingClassifier
```

```python
voting_clf=VotingClassifier(estimators=[('ex',extra),('xgb',xgb),('rf',rf),('svc',svc)],voting='soft')
```

```python
voting_clf.fit(X_train,y_train)
```

```python
for model in (extra, xgb, rf, voting_clf):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(model.__class__.__name__, f1_score(y_test, y_pred)) 
```

```python
pred_1=extra.predict(X_test)
```

```python
f1_score(y_test,pred_1,pos_label=1,average='binary')
```
